```python
regex = r"gpio_set_mask\((\d+)\);\ngpio_clr_mask\((\d+)\);"
```

```python
pin_data = re.findall(regex,data)
pin_data[:10]
```

```python
last = 67

sol = chr(last)

for set_pins,clear_pins in pin_data[1:]:
    
    """
    gpio_set_mask => OR current pins with the mask to set new pins and keep existing ones
    gpio_clr_mask => CLEAR every pin where the mask has a 1 => AND with 
    """
    
    last = last | int(set_pins) # setting pins is the same as OR
    last = last & (~int(clear_pins)) # clearing a pin is the same as applying an inverted AND
    
    sol += chr(last)

sol
```
